# 06 — Policy Brief: One-Pager Generator

Generates a shareable single-page policy brief as a high-resolution PNG and PDF.

All statistics independently verified. Every claim sourced to HHSC, BLS, or LBB data.

```python
import matplotlib
matplotlib.rcParams['text.parse_math'] = False  # disable math mode — we have $ in text

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import numpy as np
import pandas as pd
from pathlib import Path

REPORTS = Path('../reports')
PROCESSED = Path('../data/processed')
```

```python
# === POLICY BRIEF: Single-page infographic ===

fig = plt.figure(figsize=(8.5, 11), facecolor='white')  # letter size
fig.subplots_adjust(left=0.08, right=0.92, top=0.95, bottom=0.03)

ax = fig.add_axes([0, 0, 1, 1])
ax.set_xlim(0, 8.5)
ax.set_ylim(0, 11)
ax.axis('off')

# --- HEADER ---
ax.fill_between([0, 8.5], 10.2, 11, color='#b71c1c', zorder=0)
ax.text(4.25, 10.72, 'THE INVISIBLE PAY CUT',
        fontsize=26, fontweight='bold', ha='center', va='center', color='white',
        fontfamily='sans-serif')
ax.text(4.25, 10.35, "How Texas's $10.60 Wage Assumption Drives a 100,000-Person Waitlist",
        fontsize=11, ha='center', va='center', color='#ffcdd2',
        fontfamily='sans-serif')

# --- FOUR HEADLINE STATS ---
stats = [
    ('$10.60/hr', 'HHSC default attendant\nwage assumption', '#d32f2f',
     'PFD Wage Calculator cell B7\n(updated Feb 7, 2025)'),
    ('26%', 'Purchasing power lost\nsince ~2014', '#e65100',
     'BLS CPI-U South Region\n$10.60 now worth $7.82'),
    ('$4.8B/yr', 'Worker "subsidy" to\nthe Medicaid system', '#1565c0',
     '314,610 workers x $15,392\nforegone wages vs. retail'),
    ('100,000+', 'Texans on the HCS\ninterest list', '#2e7d32',
     '87 years to clear at\ncurrent funding pace'),
]

stat_y = 9.5
for i, (number, label, color, source) in enumerate(stats):
    x = 1.1 + i * 1.7
    rect = mpatches.FancyBboxPatch((x - 0.7, stat_y - 0.55), 1.4, 1.1,
                                   boxstyle='round,pad=0.1',
                                   facecolor=color, alpha=0.1,
                                   edgecolor=color, linewidth=1.5)
    ax.add_patch(rect)
    ax.text(x, stat_y + 0.25, number, fontsize=18, fontweight='bold',
            ha='center', va='center', color=color, fontfamily='sans-serif')
    ax.text(x, stat_y - 0.15, label, fontsize=7, ha='center', va='center',
            color='#333333', fontfamily='sans-serif')
    ax.text(x, stat_y - 0.45, source, fontsize=5.5, ha='center', va='center',
            color='#888888', fontfamily='sans-serif', style='italic')

# --- DIVIDER ---
ax.plot([0.5, 8.0], [8.7, 8.7], color='#cccccc', linewidth=0.5)

# --- WAGE COMPARISON CHART ---
chart_y_start = 7.0
ax.text(4.25, 8.55, 'WHAT TEXAS PAYS CARE WORKERS vs. MARKET ALTERNATIVES',
        fontsize=10, fontweight='bold', ha='center', va='center', color='#333333')

wages = [
    ("Buc-ee's", 18.00, '#388e3c'),
    ('Amazon', 17.50, '#388e3c'),
    ('H-E-B', 15.00, '#388e3c'),
    ('Walmart', 14.00, '#388e3c'),
    ('HHSC "raise"', 13.00, '#f57c00'),
    ('BLS TX HHA mean', 12.19, '#1976d2'),
    ('HHSC assumption', 10.60, '#d32f2f'),
    ('TX minimum', 7.25, '#757575'),
]

bar_height = 0.17
chart_left = 2.2
chart_width = 5.5
max_wage = 20

for i, (employer, wage, color) in enumerate(wages):
    y = chart_y_start + (len(wages) - 1 - i) * 0.21
    bar_w = (wage / max_wage) * chart_width
    rect = mpatches.FancyBboxPatch((chart_left, y - bar_height/2), bar_w, bar_height,
                                   boxstyle='round,pad=0.02',
                                   facecolor=color, alpha=0.8, edgecolor='none')
    ax.add_patch(rect)
    ax.text(chart_left - 0.1, y, employer, fontsize=7, ha='right', va='center',
            color='#333333', fontfamily='sans-serif')
    ax.text(chart_left + bar_w + 0.1, y, f'${wage:.2f}', fontsize=7, ha='left',
            va='center', fontweight='bold', color=color, fontfamily='sans-serif')

# --- DIVIDER ---
ax.plot([0.5, 8.0], [6.8, 6.8], color='#cccccc', linewidth=0.5)

# --- THE SYSTEM ---
ax.text(4.25, 6.55, 'THE SYSTEM', fontsize=12, fontweight='bold',
        ha='center', va='center', color='#b71c1c')

system_lines = [
    "HHSC rate methodology assumes attendants earn $10.60/hr -- frozen since ~2014,",
    "now worth just $7.82 in real terms (26% erosion). This caps what providers can pay,",
    "driving workers to retail jobs paying $5-8/hr more. The resulting 50% annual turnover",
    "costs ~$897M/yr statewide. Providers who cannot staff homes stop accepting new clients,",
    "feeding a 100,000+ person waitlist that would take 87 years to clear. Over a 30-year",
    "career, a direct care worker earns $624,000 less than a Buc-ee's cashier.",
]
for i, line in enumerate(system_lines):
    ax.text(4.25, 6.25 - i * 0.16, line, fontsize=7.5, ha='center', va='center',
            color='#333333', fontfamily='sans-serif')

# --- DIVIDER ---
ax.plot([0.5, 8.0], [5.2, 5.2], color='#cccccc', linewidth=0.5)

# --- CAUSAL CHAIN ---
chain_y = 4.8
chain_items = [
    ('$10.60\nfrozen', '#d32f2f'),
    ('Rates\ncapped', '#f57c00'),
    ('Workers\nleave', '#1976d2'),
    ('50%\nturnover', '#9c27b0'),
    ('Homes\nclose', '#757575'),
    ('100K+\nwait', '#d32f2f'),
]

for i, (text, color) in enumerate(chain_items):
    x = 1.0 + i * 1.2
    circle = plt.Circle((x, chain_y), 0.35, facecolor=color, alpha=0.15,
                        edgecolor=color, linewidth=1.5)
    ax.add_patch(circle)
    ax.text(x, chain_y, text, fontsize=7, fontweight='bold', ha='center', va='center',
            color=color)
    if i < len(chain_items) - 1:
        ax.annotate('', xy=(x + 0.5, chain_y), xytext=(x + 0.35, chain_y),
                    arrowprops=dict(arrowstyle='->', color='#999999', lw=1.5))

# --- DIVIDER ---
ax.plot([0.5, 8.0], [4.2, 4.2], color='#cccccc', linewidth=0.5)

# --- THE ASK ---
ax.fill_between([0.5, 8.0], 3.3, 4.1, color='#1b5e20', alpha=0.08)
ax.text(4.25, 3.95, 'THE ASK', fontsize=12, fontweight='bold',
        ha='center', va='center', color='#1b5e20')

ask_lines = [
    "Fund a $15/hr attendant wage floor in the 90th Legislature General Appropriations Act.",
    "At $15/hr, the wage matches inflation-adjusted parity ($14.37) and approaches the",
    "HCS/ICF house breakeven (~$15.50). Estimated biennial cost: ~$4.5B all funds",
    "(~$1.8B GR at 40.6% FMAP). This is less than the $4.8B workers already forfeit annually.",
]
for i, line in enumerate(ask_lines):
    ax.text(4.25, 3.75 - i * 0.13, line, fontsize=7.5, ha='center', va='center',
            color='#1b5e20', fontfamily='sans-serif')

# --- DIVIDER ---
ax.plot([0.5, 8.0], [3.15, 3.15], color='#cccccc', linewidth=0.5)

# --- SOURCES ---
ax.text(4.25, 2.95, 'SOURCES', fontsize=8, fontweight='bold',
        ha='center', va='center', color='#666666')

sources = [
    'HHSC PFD Personal Attendant Base Wage Calculator (pfd.hhs.texas.gov, updated 2/7/2025)',
    'BLS Occupational Employment and Wage Statistics, Texas, May 2024 (SOC 31-1120)',
    'BLS Consumer Price Index, South Region (CUUR0300SA0), 2015-2025',
    'LBB Article II Rider Packets, 83rd-88th Legislatures',
    'NCI Staff Stability Survey, 2023 (nationalcoreindicators.org)',
    'KERA News, "$12/hr not enough for disability services" (2025-03-04)',
    'Texas Tribune, "Low wages worsen caregiver shortage" (2024-06-17)',
]

for i, src in enumerate(sources):
    ax.text(4.25, 2.72 - i * 0.14, src, fontsize=5.5, ha='center', va='center',
            color='#999999', fontfamily='sans-serif')

# --- FOOTER ---
ax.plot([0.5, 8.0], [1.6, 1.6], color='#cccccc', linewidth=0.5)
ax.text(4.25, 1.35, 'Analysis by Christopher Ortiz  |  Data Science Investigation  |  March 2026',
        fontsize=7, ha='center', va='center', color='#aaaaaa', fontfamily='sans-serif')
ax.text(4.25, 1.15, 'All data from publicly available government sources',
        fontsize=6, ha='center', va='center', color='#bbbbbb', fontfamily='sans-serif',
        style='italic')

# --- SAVE ---
plt.savefig(REPORTS / 'policy_brief.png', dpi=200, bbox_inches='tight',
            facecolor='white', edgecolor='none')
plt.savefig(REPORTS / 'policy_brief.pdf', bbox_inches='tight',
            facecolor='white', edgecolor='none')
plt.show()

print('Saved: reports/policy_brief.png (high-res)')
print('Saved: reports/policy_brief.pdf')
```
