# 02 — House P&L Model

**Goal:** Build a per-house profit & loss model showing revenue vs. actual operating costs.

**Revenue Side:**
- ICF/IID: per-diem × beds × occupancy (1 TAC §355.456)
- HCS: unit-based fee schedule × authorized hours/units

**Cost Side:**
- Direct care staffing (24/7 coverage math)
- Overtime (the margin killer)
- Payroll taxes, workers comp, benefits
- Food, utilities, vehicles, insurance
- Admin/compliance (QIDP, documentation, audits)

**Key Insight:** "4 staff per house" is almost always understaffed for 24/7. The gap becomes overtime.

## Staffing Cost Model

24/7 coverage reality check.

```python
model = StaffingModel(
    residents=4,
    staff_per_shift=1,  # minimum 1 staff on duty 24/7
    hourly_wage=13.00,  # current Texas attendant wage assumption
)

costs = model.annual_labor_cost(overtime_hours_pct=0.15)  # 15% OT is realistic

print(f"FTEs needed: {costs['ftes']:.1f}")
print(f"Regular pay: ${costs['regular_pay']:,.0f}")
print(f"Overtime pay: ${costs['overtime_pay']:,.0f}")
print(f"Total labor cost: ${costs['total_labor_cost']:,.0f}")
```
